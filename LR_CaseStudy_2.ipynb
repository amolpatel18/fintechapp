### Check Overfitting/Underfitting

### Conclusion : Train and Test Data R-square is matching (99% and 98%) so the model is NOT overfitting.

### if R2 score on test-data is 0.99 then does it mean overfitting?

Not necessarily. An **R² score of 0.99 on test data** usually indicates an excellent fit, but whether it means **overfitting** depends on context:

---

### ✅ When it’s NOT overfitting:
- If your **training R²** is similar (e.g., 0.98–0.99) and test R² is also high → model generalizes well.
- Data is clean, features are strong predictors, and the relationship is truly linear.

---

### ✅ When it COULD be overfitting:
- If **training R² = 1.0** and test R² drops significantly (e.g., 0.99 → 0.70 on unseen data).
- If the dataset is very small or has **leakage** (test data contains info from training).
- If you used too many features relative to sample size.

---

### ✅ How to check:
1. Compare **train vs test R²**.
2. Use **cross-validation**:
   ```python
   from sklearn.model_selection import cross_val_score
   scores = cross_val_score(model, X, y, cv=5, scoring='r2')
   print(scores.mean())
   ```
3. Check **residual plots** for patterns.
4. Ensure **no data leakage**.
